```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('bank_customer_survey.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['contact','poutcome'],inplace=True)
```

```python
df.groupby('education').size()
```

```python
df.education.replace({'unknown':df.education.mode()[0]},inplace=True)
```

```python
df.groupby('job').size()
```

```python
df.job.replace({'unknown':df.job.mode()[0]},inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
x=df.drop('y',axis=1).values
y=df['y'].values
```

```python
labelencoder=LabelEncoder()
x[:,4]=labelencoder.fit_transform(x[:,4])
x[:,6]=labelencoder.fit_transform(x[:,6])
x[:,7]=labelencoder.fit_transform(x[:,7])
```

```python
pd.DataFrame(x)
```

```python
columntransform=ColumnTransformer([('encoder',OneHotEncoder(),[1,2,3,9])],remainder='passthrough')
x=columntransform.fit_transform(x)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
forest=RandomForestClassifier()
svm=SVC()
knn=KNeighborsClassifier()
tree=DecisionTreeClassifier()
```

```python
forest.fit(x_train,y_train)
svm.fit(x_train,y_train)
knn.fit(x_train,y_train)
tree.fit(x_train,y_train)
```

```python
pred_forest=forest.predict(x_test)
pred_svm=svm.predict(x_test)
pred_knn=knn.predict(x_test)
pred_tree=tree.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_tree))
```

```python
print(classification_report(y_test,pred_forest))
```

```python
print(classification_report(y_test,pred_knn))
```

```python
print(classification_report(y_test,pred_svm))
```
