```python
df=df.drop(columns=['level_0','Standard Error for Crude Rate','Year','State','Race and Hispanic Origin','index'])
```

```python
df.head()
```

```python
x=df.drop('Deaths',axis=1).values
y=df['Deaths'].values
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
labelencoder=LabelEncoder()
x[:,0]=labelencoder.fit_transform(x[:,0])
```

```python
x
```

```python
columntranform=ColumnTransformer([('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=columntranform.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
forest=RandomForestRegressor(n_estimators=20)
svm=SVR(kernel='linear')
```

```python
forest.fit(x_train,y_train)
```

```python
pred=forest.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```
