```python
columntransformer=ColumnTransformer([('enocder',OneHotEncoder(),[1,3,6,12,14,18,19])],remainder='passthrough')
x=columntransformer.fit_transform(x)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
import numpy as np
```

```python
x=sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
principle_component=pca.fit_transform(x)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

```python
# taking 35 principle component
pca=PCA(n_components=35)
new_data=pca.fit_transform(x)
```

```python
pd.DataFrame(new_data)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(new_data,y,test_size=0.25)
```

```python
os=SMOTETomek()
x_train_ns,y_train_ns=os.fit_sample(x_train,y_train)
```
